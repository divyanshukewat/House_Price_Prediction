```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
data2[~data2['total_sqft'].apply(is_float)].head(10)
```

```python
#some values are present in range therefore we will take avg and those values
# having different unit we will remove that as it won't affect our model that much
```

```python
def convert_sqft_to_num(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('2166')
```

```python
convert_sqft_to_num('2100-2850')
```

```python
x=convert_sqft_to_num('34.46Sq. Meter')
print(x)
```

```python
#converting total sqft(string) to integer 
# for range - avg is taken
data4=data2.copy()
data4['total_sqft'] = data4['total_sqft'].apply(convert_sqft_to_num)
data4.head(10)
```

```python
#checking range in converted into integer or not
# (categorical --> discrete)
#before this function total_sqft[30] was 2100-2850
data4.loc[30]
```

```python
#it is having so many categorical value we have to reduce dimension
#checking how many data points are available
data5.location=data5.location.apply(lambda x:x.strip())
location_stats=data5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
#as there are many location have less than 10 then we will sort that
len(location_stats[location_stats<10]) #these are location having less than 10 Houses.
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
data5.head()
```

```python
data5.location = data5.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(data5.location.unique())
```
